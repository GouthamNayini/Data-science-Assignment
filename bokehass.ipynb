Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

Glyphs in Bokeh are the visual markers used to represent data on a plot. Examples of glyphs include circles, squares, lines, and more. You can add glyphs to a Bokeh plot using functions like figure.scatter(), figure.line(), etc.

```python
from bokeh.plotting import figure, show


x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 9]
p = figure(title="Circle Glyph Example", x_axis_label="X-axis", y_axis_label="Y-axis")

p.circle(x, y, size=10, color="red")

show(p)
```

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in

```python
from bokeh.models import ColumnDataSource
from bokeh.io import curdoc
import numpy as np

# Create some sample data
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x, y=y))

# Create a Bokeh figure
p = figure(title="Real-Time Sin Wave", x_axis_label="X-axis", y_axis_label="Y-axis")
line = p.line('x', 'y', source=source)

# Update function for real-time updates
def update():
    new_y = np.sin(x + curdoc().session_context.request_time())
    source.data = dict(x=x, y=new_y)

# Add periodic callback to update the plot every 100 milliseconds
curdoc().add_periodic_callback(update, 100)

# Show the plot
show(p, notebook_handle=True)
```

Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?

```python
from flask import Flask, render_template
from bokeh.embed import components
from bokeh.plotting import figure
from bokeh.resources import CDN

app = Flask(__name__)

@app.route('/')
def index():
    # Create a Bokeh figure
    p = figure(title="Embedded Bokeh Plot", x_axis_label="X-axis", y_axis_label="Y-axis")
    p.circle([1, 2, 3], [4, 5, 6])

    # Embed the plot components into the HTML template
    script, div = components(p, CDN)

    # Render the template with the embedded plot components
    return render_template('index.html', script=script, div=div)

if __name__ == '__main__':
    app.run(debug=True)
```
