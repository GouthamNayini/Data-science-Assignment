Q1. Explain GET and POST methods.

GET Method:

GET is one of the HTTP methods used in web development.
It is used for requesting data from a specified resource, typically by including parameters in the URL.
GET requests are idempotent, meaning making the same GET request multiple times should have the same result.

POST Method:

POST is another HTTP method used for submitting data to be processed to a specified resource.
It is commonly used when you want to create or update data on the server, such as submitting a form on a website.
POST requests are not idempotent, as making the same request multiple times may have different results (e.g., creating multiple records).
Data is sent in the request body, which can be in various formats (e.g., JSON, form data).

In Flask, the request object is used to access data sent with an HTTP request. It allows you to retrieve information such as form data, URL parameters, headers, and more from the incoming request. You can use the request object to interact with the data sent by the client, process it, and make decisions or perform actions based on that data. Q3. Why is redirect() used in Flask?

The redirect() function in Flask is used to redirect the client's web browser to a different URL. It is commonly used in web applications to perform actions such as sending the user to a different page after a form submission, login, or other user interactions. Redirecting helps maintain a clean and user-friendly URL structure and can improve the overall user experience.

Q4. What are templates in Flask? Why is the render_template() function used?

Templates in Flask are HTML files that contain placeholders for dynamic content. These placeholders are usually populated with data from your Flask application. Templating allows you to separate the presentation (HTML) from the logic (Python code) of your web application, making it easier to maintain and modify your web pages.

The render_template() function in Flask is used to render HTML templates and pass data to them. It takes the name of the HTML template as an argument and can also pass variables to the template. This function combines the template with the provided data to generate the final HTML page that is sent to the client's web browser.

@app.route('/postman_action',methods=['POST'])
def math_ops1():
    if(request.method == 'POST'):
        ops = request.json['operation']
        num1 = int(request.json['num1'])
        num2 = int(request.json['num2'])
        if ops == 'add':
            r = num1+num2
            result = "The sum of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
        if ops == 'subtract':
            r = num1-num2
            result = "The subtract of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
        if ops == 'multiply':
            r = num1*num2
            result = "The multiply of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
        if ops == 'divide':
            r = num1/num2
            result = "The divide of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
            
        return jsonify(result)
