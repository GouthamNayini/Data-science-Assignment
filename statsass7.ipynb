Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation
of 5 using Python. Interpret the results.

```python
import scipy.stats as stats
mean = 50
std_dev = 5
sample_size = 30 
standard_error = std_dev / (sample_size**0.5)
confidence_level = 0.95
margin_of_error = stats.norm.ppf((1 + confidence_level) / 2) * standard_error
confidence_interval = (mean - margin_of_error, mean + margin_of_error)
print("95% Confidence Interval:", confidence_interval)
```

Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag
matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20%
brown. Use Python to perform the test with a significance level of 0.05.

```python
import pandas as pd
from scipy import stats
expected_colors = [0.2, 0.2, 0.2, 0.1, 0.1, 0.2]
colors = ["blue"] * 20 + ["orange"] * 20 + ["green"] * 20 + ["yellow"] * 10 + ["red"] * 10 + ["brown"] * 20
observed_colors = pd.Series(colors).value_counts()
chi2, pval, dof, expected = stats.chi2_contingency([observed_colors, expected_colors])
alpha = 0.05 
if pval < alpha:
    print("Reject the null hypothesis. The observed distribution of colors is statistically different from the expected distribution.")
else:
    print("Fail to reject the null hypothesis. The observed distribution of colors may be due to chance.")
```

Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals
smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the
population who smoke.

```python
import numpy as np
n = 500
smokers = 60
confidence_level = 0.95
p_hat = smokers / n
standard_error = np.sqrt(p_hat * (1 - p_hat) / n)
margin_of_error = standard_error * 1.96 
lower_bound = p_hat - margin_of_error
upper_bound = p_hat + margin_of_error
print(f"95% confidence interval for the proportion of smokers: [{lower_bound:.4f}, {upper_bound:.4f}]")
```

Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the
area corresponding to a chi-square statistic of 15.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as statsm
degrees_of_freedom = 10
x_values = np.linspace(0, 30, 1000)
plt.plot(x_values, stats.chi2.pdf(x_values, degrees_of_freedom), label=f'Chi-square Distribution (df={degrees_of_freedom})')
x_shaded = np.linspace(15, 30, 1000)
plt.fill_between(x_shaded, stats.chi2.pdf(x_shaded, degrees_of_freedom), color='orange', alpha=0.5, label='Chi-square Statistic = 15')
plt.title('Chi-square Distribution')
plt.xlabel('Chi-square Statistic')
plt.ylabel('Probability Density Function')
plt.legend()
plt.show()
```

Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520
preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who
prefer Coke.

```python
import scipy.stats as stats
import math
sample_size = 1000
prefer_coke = 520
sample_proportion = prefer_coke / sample_size
confidence_level = 0.99
z_score = stats.norm.ppf((1 + confidence_level) / 2)
margin_of_error = z_score * math.sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)
confidence_interval = (sample_proportion - margin_of_error, sample_proportion + margin_of_error)
print("99% Confidence Interval:", confidence_interval)
```

Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe
45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the
expected frequencies of a fair coin. Use a significance level of 0.05.

```python
import numpy as np
from scipy.stats import chi2_contingency
observed_tails = 45
total_flips = 100
expected_heads = total_flips / 2
expected_tails = total_flips / 2
observed_frequencies = np.array([total_flips - observed_tails, observed_tails])
expected_frequencies = np.array([expected_heads, expected_tails])
chi2_stat, p_value, dof, expected = chi2_contingency([observed_frequencies])
print("Chi-square Statistic:", chi2_stat)
print("P-value:", p_value)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:", expected)
alpha = 0.05
print("\nSignificance Test:")
if p_value < alpha:
    print("Reject the null hypothesis. There is evidence that the coin is biased towards tails.")
else:
    print("Fail to reject the null hypothesis. There is no significant evidence that the coin is biased towards tails.")
```

Q11. A random sample of 30 people was selected from a population with an unknown mean and standard
deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10.
Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a
significance level of 0.05.

```python
import scipy.stats

sample_mean = 72
sample_std = 10
sample_size = 30
null_hypothesis_mean = 70
t_statistic = (sample_mean - null_hypothesis_mean) / (sample_std / (sample_size**0.5))

degrees_of_freedom = sample_size - 1
alpha = 0.05
critical_value = scipy.stats.t.ppf(1 - alpha/2, degrees_of_freedom)
print("t-statistic:", t_statistic)
print("Critical value:", critical_value)

if abs(t_statistic) > critical_value:
    print("Reject the null hypothesis. ")
else:
    print("Fail to reject the null hypothesis. ")
```
